```python
import pandas as pd
import os
```

```python
# find files
file_paths = ["Resource/JC-202303-citibike-tripdata.csv","Resource/JC-202304-citibike-tripdata.csv","Resource/JC-202305-citibike-tripdata.csv"]

# List to hold dataframes for each CSV file
dataframes = []

# Read each CSV file into a dataframe and append to the list
for file_path in file_paths:
    df = pd.read_csv(file_path)
    dataframes.append(df)

# Concatenate all dataframes into one
merged_df = pd.concat(dataframes)
```

```python
merged_df.head()
```

```python
merged_df.info()
```

```python
# Convert "started_at" and "ended_at" to datetime data type
merged_df['started_at'] = pd.to_datetime(merged_df['started_at'])
merged_df['ended_at'] = pd.to_datetime(merged_df['ended_at'])

# Create new columns for month and time
merged_df['start_month'] = merged_df['started_at'].dt.month
merged_df['end_month'] = merged_df['ended_at'].dt.month

# Assuming you want time in hours and minutes
merged_df['start_time'] = merged_df['started_at'].dt.strftime('%H:%M')
merged_df['end_time'] = merged_df['ended_at'].dt.strftime('%H:%M')

merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df['end_time'] = pd.to_datetime(merged_df['end_time'])
```

```python
merged_df['start_time'] = pd.to_datetime(merged_df['start_time'])
```

```python
merged_df.info()
```

```python
merged_df['duration'] = merged_df['end_time'] - merged_df['start_time']
```

```python
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df['duration'] = merged_df['duration'].dt.seconds / 3600  # Convert seconds to hours
```

```python
merged_df.head()
```

```python
merged_df['duration'] = merged_df['duration'].round(2)
```

```python
merged_df.head()
```

```python
# Drop specified columns
columns_to_remove = ["end_month"]
merged_df = merged_df.drop(columns=columns_to_remove)

# Display the updated DataFrame
merged_df.head()
```

```python
# Define a function to map time values to period labels
def map_time_to_period(time):
    if time < pd.Timestamp("06:00:00").time():
        return "early morning"
    elif time < pd.Timestamp("12:00:00").time():
        return "morning"
    elif time < pd.Timestamp("18:00:00").time():
        return "afternoon"
    else:
        return "night"

# Create the "Period of the day" column based on the "start_time" column
merged_df['Period of the day'] = merged_df['start_time'].dt.time.apply(map_time_to_period)
```

```python
# Group by 'start_station_name' and select the first occurrence of 'start_lat' and 'start_lng' in each group
start_lat_lng = merged_df.groupby('start_station_name')[['start_lat', 'start_lng']].first()

# Replace missing values in 'start_station_name' with an empty string
merged_df['start_station_name'].fillna('', inplace=True)

# Replace 'start_lat' and 'start_lng' with the selected values for each respective 'start_station_name'
merged_df[['start_lat', 'start_lng']] = merged_df.apply(lambda row: start_lat_lng.loc[row['start_station_name']] if row['start_station_name'] != '' else [pd.NA, pd.NA], axis=1)

merged_df.tail()
```
